```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import pandas as pd
import pickle

TOTAL_EPISODES = 2000
```

```
with open ('MC_episode_scores', 'rb') as fp:
    MC_episode_scores = pickle.load(fp)

with open ('MC_bonus_scores', 'rb') as fp:
    MC_bonus_scores = pickle.load(fp)
```

```
with open ('Off_MC_episode_scores', 'rb') as fp:
    off_MC_episode_scores = pickle.load(fp)

with open ('Off_MC_bonus_scores', 'rb') as fp:
    off_MC_bonus_scores = pickle.load(fp)
```

```
with open ('sarsa_episode_scores', 'rb') as fp:
    sarsa_episode_scores = pickle.load(fp)

with open ('sarsa_bonus_scores', 'rb') as fp:
    sarsa_bonus_scores = pickle.load(fp) 
```

```
with open ('DQ_episode_scores', 'rb') as fp:
    DQ_episode_scores = pickle.load(fp)

with open ('DQ_bonus_scores', 'rb') as fp:
    DQ_bonus_scores = pickle.load(fp)
```

```
with open ('3_TB_episode_scores', 'rb') as fp:
    TB3_episode_scores = pickle.load(fp)


with open ('3_TB_bonus_scores', 'rb') as fp:
    TB3_bonus_scores = pickle.load(fp)
```

```
with open ('nstep_expsarsa_episode_scores', 'rb') as fp:
    nstep_expsarsa_episode_scores = pickle.load(fp)

with open ('nstep_expsarsa_bonus_scores', 'rb') as fp:
    nstep_expsarsa_bonus_scores = pickle.load(fp)
```

```
plt.figure(figsize=(25, 10))
plt.plot(range(TOTAL_EPISODES), MC_episode_scores, color = 'rebeccapurple', label = 'Monte Carlo - On Policy')
#plt.plot(range(TOTAL_EPISODES), off_MC_episode_scores, color = '#D0D3D4', label = 'Monte Carlo - Off Policy')
plt.plot(range(TOTAL_EPISODES), sarsa_episode_scores, color = 'thistle', label = 'SARSA')
plt.plot(range(TOTAL_EPISODES), DQ_episode_scores, color = '#E6C5F1', label = 'Double Q-learning')
plt.plot(range(TOTAL_EPISODES), TB3_episode_scores, color = '#633974', label = '3-Step Backup-Tree')
plt.plot(range(TOTAL_EPISODES), nstep_expsarsa_episode_scores, color = '#FFF1FF', label = '2-Step Expected SARSA')
plt.xlabel('episodes ->')
plt.ylabel('epsiode score ->')
plt.legend()
plt.show()
```

```
plt.figure(figsize=(25, 10))
window_size = 50
MC_episode_scores_smooth = pd.Series(MC_episode_scores).rolling(window_size , min_periods = window_size).mean()
Off_MC_episode_scores_smooth = pd.Series(off_MC_episode_scores).rolling(window_size , min_periods = window_size).mean()
sarsa_episode_scores_smooth = pd.Series(sarsa_episode_scores).rolling(window_size , min_periods = window_size).mean()
DQ_episode_scores_smooth = pd.Series(DQ_episode_scores).rolling(window_size , min_periods = window_size).mean()
TB3_episode_scores_smooth = pd.Series(TB3_episode_scores).rolling(window_size , min_periods = window_size).mean()
nstep_expsarsa_episode_scores_smooth = pd.Series(nstep_expsarsa_episode_scores).rolling(window_size , min_periods = window_size).mean()


plt.plot(MC_episode_scores_smooth, color = '#AE8FB9', label = 'Monte Carlo - on policy')
#plt.plot(Off_MC_episode_scores_smooth, color = '#D0D3D4', label = 'Monte Carlo - Off Policy')
plt.plot(sarsa_episode_scores_smooth, color = 'thistle', label = 'SARSA')
plt.plot(DQ_episode_scores_smooth, color = '#E6C5F1', label = 'Double Q-learning')
plt.plot(TB3_episode_scores_smooth, color = '#633974', label = '3-Step Backup-Tree')
plt.plot(nstep_expsarsa_episode_scores_smooth, color = '#A97DB9', label = '2-Step Expected SARSA')
plt.xlabel("epsiode ->")
plt.ylabel("epsiode score (smoothed) -> ")

plt.legend()
plt.show()
```

```
algorithms = ['On-Policy Monte Carlo', 'Off-Policy Monte Carlo', 'SARSA', 'Double Q-learning', '2-Step Expected SARSA', '3-Step TreeBackup']
scores = [MC_episode_scores, off_MC_episode_scores, sarsa_episode_scores, DQ_episode_scores, TB3_episode_scores, nstep_expsarsa_episode_scores]
for i in range(len(algorithms)): print("\n{} : median(last 100) = {}, variance(last 100) = {}".format(algorithms[i], np.median(scores[i][-100:]), np.round(np.var(scores[i][-100:]), 2)))
```

```
bonus_scores = [MC_bonus_scores, off_MC_bonus_scores, sarsa_bonus_scores, DQ_bonus_scores, TB3_bonus_scores, nstep_expsarsa_bonus_scores]
for i in range(len(algorithms)): 
  print("\n{} : max reward = {}, median(last 100) = {}, variance(last 100) = {}".format(algorithms[i], np.max(bonus_scores[i]),np.median(bonus_scores[i][-100:]), np.round(np.var(bonus_scores[i][-100:]), 2) ))
```
